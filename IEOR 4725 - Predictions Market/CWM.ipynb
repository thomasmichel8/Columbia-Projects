```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('training.csv')
```

```python
df
```

```python
dfGDP = df.loc[df['INDICATOR'] == "RealGDP"].fillna(0)
```

```python
dfGDP = dfGDP[['INDICATOR','QUARTER','FORECASTER ID', 'YEAR BEING FORECAST','actual','pred_average','pred_var']]
```

```python
y = pd.get_dummies(dfGDP['FORECASTER ID'], prefix='FORECASTER')
dfGDP = dfGDP.join(y)
dfGDP = dfGDP.drop('FORECASTER ID',axis = 1)
dfGDP = dfGDP.drop('INDICATOR',axis = 1)
dfGDP
```

```python
dfGDP
```

```python
dfGDP = dfGDP[dfGDP.actual != 0]
```

```python
np.random.seed(0)
rand = dfGDP.sample(frac = 1)
#80/20 split
nrow = len(rand)
train_prop = int(np.round(nrow*0.8))
train = rand[:train_prop]
val = rand[train_prop:]
```

```python
train.loc[:, train.columns != 'actual']
```

### Random Forest Model

```python
X_train, Y_train = train.loc[:, train.columns != 'actual'],train[['actual']]
val_X, val_Y =  val.loc[:, train.columns != 'actual'],val[['actual']]
```

```python
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(max_depth=8, random_state=0)
regr.fit(X_train, Y_train)
```

```python
from sklearn.metrics import mean_squared_error
preds = regr.predict(val_X)

RF_MSE = mean_squared_error(val_Y,preds)
SPF_MSE = mean_squared_error(val_Y,np.array(val_X['pred_average'].tolist()))

print('RF_MSE',RF_MSE)
print('SPF_MSE',SPF_MSE)
```

###  MLP Model

```python
from sklearn.neural_network import MLPRegressor

mlp = MLPRegressor(random_state=1, max_iter=1000).fit(X_train, Y_train)
```

```python
preds = mlp.predict(val_X)

RF_MSE = mean_squared_error(val_Y,preds)
SPF_MSE = mean_squared_error(val_Y,np.array(val_X['pred_average'].tolist()))

print('RF_MSE',RF_MSE)
print('SPF_MSE',SPF_MSE)
```

## CWM

```python
dfGDP = df.loc[df['INDICATOR'] == "RealGDP"].fillna(0)
dfGDP = dfGDP[['INDICATOR','QUARTER', 'FORECASTER ID', 'YEAR BEING FORECAST','actual','pred_average','pred_var']]
dfGDP = dfGDP.drop('INDICATOR',axis = 1)
dfGDP = dfGDP[dfGDP.actual != 0]
dfGDP
```

```python
#new col:actual bin check if pre_avg droped in the actual gdp bin.
dfGDP['bin_check'] = np.where( (dfGDP['pred_average'] <= (dfGDP['actual'] + 0.5)) 
                              & (dfGDP['pred_average'] >= (dfGDP['actual']-0.5)), 1, 0)

dfGDP = dfGDP[dfGDP['actual']!= 0]
dfGDP.describe()
```

```python
dfGDP.sort_values(by=['FORECASTER ID'])
```

```python
import math

list_of_forecaster= []
list_of_Contribution = []
for i in dfGDP['FORECASTER ID'].unique():
    list_of_forecaster.append(i)
    
    
    list_Si_minus_Si_without_j = []
    total_events_predicted = 0
    
    for j in dfGDP['YEAR BEING FORECAST'].unique():
        
        
        
        for k in range(1,5):
            
            
            O_ir_minus_M_ir_square = dfGDP[dfGDP['YEAR BEING FORECAST'] == j][dfGDP['QUARTER'] == k]['bin_check'].mean()
            
            si = 100 - 50 * ((1 - O_ir_minus_M_ir_square)**2 
                           + (0 - (1 - O_ir_minus_M_ir_square) )**2)
            
            
            sj = dfGDP[dfGDP['YEAR BEING FORECAST'] == j][dfGDP['QUARTER'] == k][dfGDP['FORECASTER ID']== i]['bin_check'].mean()
            if math.isnan(sj) == True:
                total_events_predicted += 0
            else:
                
                count = dfGDP[dfGDP['YEAR BEING FORECAST'] == j][dfGDP['QUARTER'] == k]['bin_check'].count()
                summ  = dfGDP[dfGDP['YEAR BEING FORECAST'] == j][dfGDP['QUARTER'] == k]['bin_check'].sum()
                
                without_j = (summ - sj)/(count-1)
                
                Si_minus_Si_without_j = 100 - 50 * ((1 - without_j ) ** 2 
                                                    +(0- (1 - without_j)) ** 2)
                list_Si_minus_Si_without_j.append(si - Si_minus_Si_without_j)
                total_events_predicted += 1
    
    Cj = sum(list_Si_minus_Si_without_j) / total_events_predicted
    list_of_Contribution.append(Cj)
```

```python
d = {'forecaster': list_of_forecaster, 'Contribution': list_of_Contribution}
df = pd.DataFrame(data=d)
df
# df['Contribution'].max()
# df[df['forecaster'] == 40]
# df[df['Contribution']==-1.374716553287982]
```

```python
gdp_prediction_list = []
actual_list = []
for j in dfGDP['YEAR BEING FORECAST'].unique():
    for k in range(1,5):
        pre_avg = dfGDP[dfGDP['YEAR BEING FORECAST'] == j][dfGDP['QUARTER'] == k]['pred_average'].mean()
        gdp_prediction_list.append(pre_avg)
        
        actual_list.append(dfGDP[dfGDP['YEAR BEING FORECAST'] == j][dfGDP['QUARTER'] == k]['actual'].mean())
```

```python
mean_squared_error(gdp_prediction_list,actual_list)
```

```python
gdp_prediction_contribution_list = []
for j in dfGDP['YEAR BEING FORECAST'].unique():
    for k in range(1,5):
        number = []
        forecasts = []
        for i in df_positive['forecaster'].unique():
            weight = df_positive[df_positive['forecaster'] == i]['Contribution'].mean()
            forecaster_i = dfGDP[dfGDP['YEAR BEING FORECAST'] == j][dfGDP['QUARTER'] == k][dfGDP['FORECASTER ID']== i]['pred_average'].mean()  

            if math.isnan(weight) or math.isnan(forecaster_i) == True:
                pass
            else:
                number.append(1)
                forecasts.append(forecaster_i)
        
        gdp_prediction_contribution_list.append(sum(forecasts) / sum(number))                                          
```

```python
mean_squared_error(gdp_prediction_contribution_list,actual_list)
```

```python
gdp_prediction_CWM_list = []
for j in dfGDP['YEAR BEING FORECAST'].unique():
    for k in range(1,5):
        weight_list = []
        forecasts = []
        for i in df_positive['forecaster'].unique():
            weight = df_positive[df_positive['forecaster'] == i]['Contribution'].mean()
            forecaster_i = dfGDP[dfGDP['YEAR BEING FORECAST'] == j][dfGDP['QUARTER'] == k][dfGDP['FORECASTER ID']== i]['pred_average'].mean()  
            if math.isnan(weight) or math.isnan(forecaster_i) == True:
                pass
            else:

                weight_list.append(weight)
                forecasts.append(forecaster_i * weight)
        
        gdp_prediction_CWM_list.append(sum(forecasts) / sum(weight_list))       
```

```python
mean_squared_error(gdp_prediction_CWM_list,actual_list)
```
